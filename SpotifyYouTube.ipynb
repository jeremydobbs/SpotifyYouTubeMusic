```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/jeremydobbs/Desktop/Code/SpotifyYouTube/Spotify_Youtube.csv')
display(df)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.drop(['Unnamed: 0','Uri','Url_youtube','Channel','official_video','Description','Url_spotify','Title','Track'],axis=1)
display(df)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
pd.options.display.float_format = '{:,.0f}'.format
views_likes_comments = df.groupby(['Artist'])[['Views','Likes','Comments']].sum().sort_values(by='Views',ascending=False)
display(views_likes_comments.head(20))
```

```python
views = df.groupby(['Artist'])['Views'].sum().reset_index().sort_values(by='Views',ascending=False).head(10)
views
```

```python
plt.title('Top 10 Views on YouTube')
ax = sns.barplot(data=views,y='Artist',x='Views',orient='h',palette='Blues')
plt.xlabel('Views')
plt.ylabel('Artist')
plt.show()
```

```python
likes = df.groupby(['Artist'])['Likes'].sum().reset_index().sort_values(by='Likes',ascending=False).head(10)
likes
```

```python
plt.title('Top 10 Likes on YouTube')
ax = sns.barplot(data=likes,y='Artist',x='Likes',orient='h',palette='Blues')
plt.xlabel('Likes')
plt.ylabel('Artist')
plt.show()
```

```python
comments = df.groupby(['Artist'])['Comments'].sum().reset_index().sort_values(by='Comments',ascending=False).head(10)
comments
```

```python
plt.title('Top 10 Comments on YouTube')
ax = sns.barplot(data=comments,y='Artist',x='Comments',orient='h',palette='Blues')
plt.xlabel('Comments')
plt.ylabel('Artist')
plt.show()
```
